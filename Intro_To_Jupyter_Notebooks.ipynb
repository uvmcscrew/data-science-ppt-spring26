## How Jupyter Notebooks Work

Jupyter notebooks are made up of different code cells that can be run separately and in any order

These cells can be set to code (the default and runs python code), markdown (which act as a markdown file), and raw (which is just raw text)

To edit a cell, just click on it if it is a code or raw cell, and double click on it if it is a markdown cell. if it is a markdown cell, to return it to the finished markdown formatting, press the run button (the '>' in the second row at the top

```python
# jupyter notebooks share variables across cells
x = 5
```

```python
x = x + 1
print(x)
```

Basically, each cell acts as a seperatly runnable python script that adds the variables declared in that script to a global enviroment
