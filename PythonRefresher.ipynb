### How to declare a variable

```python
x = 5 #declares the variable x and sets it equal to 5. you can't just declare a variable
print(x)
x = 'hello' #sets x equal to 'hello'
print(x)
```

notice how x changed from 5 to hello. this is because python is a dynamically typed language, and does not have type checking

### What is a list, and how to use them

lists are a neat and helpful way of containing a bunch of values. these can be values stored by variables, new values added to the list, and more. they can be looped for, which you will see in the next section, and overall are a great way of storing similar bits of information together

the code chunks below will show some simple things that manipulate lists, they are by no means extensive though. if you want a list to do something, it probably can

```python
# this creates an empty list
list_1 = []
print(list_1)
```

```python
# this adds the value 2 to the end of the list
list_1.append(2)
print(list_1)
# this now adds the value 'hi' to the end of the list
list_1.append('hi')
print(list_1)
# adding the value of a variable
random_value = '231231'
list_1.append(random_value)
print(list_1)
```

notice how the list can have values of multiple types. this is because lists are not required only have elements from the same type, though most lists do

```python
# this gets rid of the value at the end of the list
list_1.pop()
print(list_1)
```

```python
# this is how you can get the length of a list
print(len(list_1))
```

```python
#to get a specific element from a list you use its location. And python is a 0 indexed language, meaning lists start at location 0
print(list_1)
#this should give us 'hi'
print(list_1[1]) #the format to get the specific value is listname[location]
```

### How to make a for loop

here is a basic for loop that just adds to an integer variable until it equals 10

```python
x = 0 #creating the number you will be adding to
for j in range(10): #this says for each element that we will call j, in the range 0 to 9, do the action inside the for loop
    x = x + 1 # adds 1 to x

print(x)
```

for loops turn out to be quite useful in python, and making graphs in general. You can loop over elements in a list, which is shown below, and this can be very useful when graphing multiple different variables, or just making multiple graphs

```python
# here is how to loop over every element in a list

# making an example list using a built in way of creating a list based off of a for loop
list_2 = [j+1 for j in range(10)] #this makes an element in the list of value j+1 for each value j in the range 0 to 9
print(list_2)

#looping over every element in that for loop
for i in list_2:
    print(i)
```

yes, it is just that simple to loop over lists. this is why lists are super neat, and why they are used so frequently

### How to use dictionaries

For so, so long I avoided dictionaries. But they are actually rather useful. See, i kept having this problem where i wanted to be able to modify the variable name I was calling inside a loop, but this is not something python lets you do. But, you can modify which element of a dictionary you are calling. This makes dictionaries ...... ACTUALLY just a list of stored variable names and their values. This is different from a list where you keep appending variables, because in that case you are only storing the values of the variables, not the names as well

```python
# Here is how you create an empty dictionary
dictionary_1 = {}
print(dictionary_1)
```

```python
# here is how to create a dictionary with values
dictionary_2 = {'variable name 2' : 1, # the string on the left is the name you access the value with. the value on the left is the value
                # the comma tells python that the next string is a new element in the dictionary
                'variable name 3' : 'hello :)'} #here is another example of a value in a dictionary
print(dictionary_2)
```

In fact, adding new elements to a dictionary is also really easy

```python
# here is how to do it
dictionary_2['new variable name'] = 'new variable value' #adds a new element if there is not one already by that name
#and assigns it the new value
print(dictionary_2)
```

```python
#you can even modify the elements already in the dictionary by just calling them similar to a variable
dictionary_2['variable name 2'] = dictionary_2['variable name 2'] + 10
print(dictionary_2)
```

now, looping through dictionaries is a bit more complicated. a dictionary has 3 seperate methods to access its information. 
- the first method is the keys method, and this returns a list (not quite but its close enough to a list) of the keys in the dictionary. Keys are just the name used to get the value contained at that location in the dictionary. If we think of the dictionary as a list of variable names and their associated values (which together we call elements), then the key is the variable name
- the second method is the values method, and this returns a list (same as above) of the values in the dictionary
- the last method is the items method, and this returns a list (same as above) of tuples containing individual key value pairs that are in the dictionary

```python
#see the keys, values, and items
print(dictionary_2.keys())
print(dictionary_2.values())
print(dictionary_2.items())
```

```python
#now here is how to loop through the dictionary
for key, value in dictionary_2.items():
    print(f'here is the key "{key}", and here is the value "{value}"')
```
